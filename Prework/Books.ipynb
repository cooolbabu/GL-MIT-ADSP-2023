```python
df1 = data[['Rating', 'Reviews', 'Book_title', 'Number_Of_Pages', 'Type', 'Price']]
df1
```

```python
df1.hist(figsize=(25,10))
```

```python
plt.figure(figsize = (15,5))
sns.countplot(x="Type", data=df1)
```

```python
plt.figure(figsize = (15,5))
sns.stripplot(y = 'Rating', x = 'Type', data = df1)
```

```python
plt.boxplot(df1['Rating'])
plt.show()
```

```python
sns.displot(df1['Reviews'], kind='kde')
plt.show()
```

```python
corr_matrix = df1.corr()
sns.heatmap(corr_matrix, annot = True)
plt.show()
```

```python
sns.heatmap(df1.corr(), annot = True)
plt.show()
```

```python
plt.pie(df1.Type.value_counts(), autopct = '%.1f%%', radius = 1.2, labels = df1.Type.unique())

# display the plot
plt.show()
```
